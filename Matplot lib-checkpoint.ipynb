```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
from scipy.stats import gaussian_kde
```

```python
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel('data.xlsx')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])# converting the datetime format
# ploting the data
plt.plot(df['Date'],df['Sales'],color = "cyan")
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=20)
plt.show()
```

```python
plt.bar(df['Product'],df['Sales'],color='cyan')
plt.title('Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
plt.scatter(df['Product'],df['Sales'],color='cyan')
# Calculate the regression line
slope, intercept = np.polyfit(df['Product'], df['Sales'], 1)
# Generate x values for the regression line
x_line = np.linspace(df['Product'].min(), df['Sales'].max(), 100)
# Calculate y values using the slope and intercept
y_line = slope * x_line + intercept
# Plot the regression line
plt.plot(x_line, y_line, color='yellow', label='Regression Line')
plt.title('scatter of the products sales ')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
car= pd.read_excel('Car data saved.xlsx')
```

```python
car.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(car['Max Price (Lakh)'],car['Insurance'],color='magenta')
# Calculate the regression line
slope, intercept = np.polyfit(car['Max Price (Lakh)'], car['Insurance'], 1)
# Generate x values for the regression line
x_line = np.linspace(car['Max Price (Lakh)'].min(), car['Max Price (Lakh)'].max(), 100)
# Calculate y values using the slope and intercept
y_line = slope * x_line + intercept
# Plot the regression line
plt.plot(x_line, y_line, color='cyan', label='Regression Line')
plt.title('Max Price (Lakh) VS Insurance ')
plt.xlabel('Max Price(Lakh)')
plt.ylabel('Insurance')
plt.grid(True)
plt.show()
```

```python
# Create the histogram
counts, bin_edge, patches = plt.hist(df['Sales'], bins=5, color='magenta', edgecolor='cyan', alpha=0.7)

# Calculate the bin centers
bin_centers = 0.5 * (bin_edge[1:] + bin_edge[:-1])

# Plot the distribution line using bin centers and normalized counts
plt.plot(bin_centers, counts, color='blue', linestyle='-', marker='o', label='Distribution Line')
plt.title('Sales per Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
