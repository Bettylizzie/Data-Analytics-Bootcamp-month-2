```python
new_df1= df.set_index(["Country","Product"])
new_df1.head()
```

## flattening ie combining two columns to be one

```python
# merging index country and product to be one column
merged_index = new_df1.index.map(lambda x: f'{x[0]}_{x[1]}')
merged_index
```

```python
new_df
```

```python
new_df.index = merged_index
new_df
```
