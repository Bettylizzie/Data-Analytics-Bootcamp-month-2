```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
car= pd.read_excel('Car data saved.xlsx')
```

```python
car.head()
```

```python
variables=['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)','CC',	'Seats','Variants',
           'Ex-Showroom Price','RTO','Insurance','Other', 'Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i,var in enumerate(variables,start=1):
    plt.subplot(4,3,i)
    sns.scatterplot(x=car[var],y=car['Max Price (Lakh)'],color='Magenta',alpha=0.7)
    plt.title(f'scatter plot of {var} vs onroad price' )
    plt.xlabel(var)
    plt.ylabel('Target')

plt.tight_layout()
plt.xticks(rotation=70)
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i,var in enumerate(variables,start=1):
    plt.subplot(3,4,i)
    sns.histplot(car[var], bins=20, kde=True, color='Magenta')
    kde_kws={'color': 'blue'}
    plt.title(f'histogram of {var}' )
    plt.xlabel(var)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
# HEAT MAP
```

```python
selected_columns=['Min Price (Lakh)','Max Price (Lakh)','Range (kmpl)','CC',	'Seats','Variants',
           'Ex-Showroom Price','RTO','Insurance','Other']
df_selected = car[selected_columns]
```
